# Figure Size And DPI

```python
fig = plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes= plt.subplots(nrows=2,ncols=1,figsize=(8,2))
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig.savefig('pic.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x**2')
ax.plot(x,x**3,label='x**3')

ax.legend()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x**2')
ax.plot(x,x**3,label='x**3')

ax.legend(loc=0)#best
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x**2')
ax.plot(x,x**3,label='x**3')

ax.legend(loc=10)#center
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label='x**2')
ax.plot(x,x**3,label='x**3')

ax.legend(loc=(0.1,0.1))
```

```python
left=pd.DataFrame({
    'key':['k0','k1','k2'],
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2']
})
```

```python
right=pd.DataFrame({
    'key':['k0','k1','k2'],
    'C':['C0','C1','C2'],
    'D':['D0','D1','D2']
})
```

```python
pd.merge(left,right)
```

```python
pd.merge(left,right,how='outer')
```

```python
pd.merge(left,right,how='left',on='key')
```

```python
pd.merge(left,right,how='right')
```

```python
left=pd.DataFrame({
    'key1':['k0','k1','k2'],
    'key2':['k1','k0','k0'],
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2']
})
```

```python
right=pd.DataFrame({
    'key1':['k0','k1','k2'],
    'key2':['k0','k0','k0'],
    'C':['C0','C1','C2'],
    'D':['D0','D1','D2']
})
```

```python
pd.merge(left,right)
```

```python
pd.merge(left,right,on='key1')
```

```python
pd.merge(left,right,on='key2')
```

```python
pd.merge(left,right,on=['key1','key2'])
```
